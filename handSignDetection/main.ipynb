Acessar as pastas com as imagens tiradas e fazer um loop em cada imagem das maos

```python
# acessar a pasta Folds_Dataset_Final
import os
import shutil
folder_path = os.path.abspath('../Folds_Dataset_Final/')

if os.path.isdir(folder_path):
  content = os.listdir(folder_path)

  for item in content:
    new_folder = item
    new_path = os.path.join(folder_path,new_folder)

    content1 = os.listdir(new_path)

    for item in content1:
      
      new_folder_2 = item
      new_path_2 = os.path.join(new_path,new_folder_2)
      print(new_path_2)
      mask = 'mask'
      maskfolder = os.path.join(new_path_2,mask)
      
      if not os.path.exists(maskfolder):
        os.mkdir(maskfolder)
      else:
        print('pasta já criada')
     
      content2 = os.listdir(new_path_2)
      
      # loop interno dentro de cada pasta para separ as imagens das mascaras
      for item in content2:
        if item.startswith('c'):
          path_origin = os.path.join(new_path_2,item)
          path_destination = os.path.join(maskfolder,item)
          shutil.move(path_origin,path_destination)
```
